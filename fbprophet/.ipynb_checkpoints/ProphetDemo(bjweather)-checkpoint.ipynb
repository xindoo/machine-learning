```python
# -*- coding: utf-8 -*-
import pandas as pd
import matplotlib.pyplot as plt
from fbprophet import Prophet
```

```python
data = pd.read_csv('bjweather.csv')
data = data.sort_values(by=['date'])
data = data.reset_index()
```

```python
dfmin = pd.DataFrame()
dfmin['ds'] = data['date']
dfmin['y'] = data['minT']
dfmax = pd.DataFrame()
dfmax['ds'] = data['date']
dfmax['y'] = data['maxT']
dfmax.head(10)
```

```python
mmin = Prophet()
mmin.fit(dfmin)
mmax = Prophet()
mmax.fit(dfmax)
```

```python
futuremin = mmin.make_future_dataframe(periods=30)
futuremax = mmax.make_future_dataframe(periods=30)
futuremin.tail()
```

```python
fcmin = mmin.predict(futuremin)
fcmax = mmax.predict(futuremax)
fcmin[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# fig0 = mmin.plot(fcmin)
fig1 = mmax.plot(fcmax)
```

```python
fig2 = mmax.plot_components(fcmin)
```

```python
fig3 = mmin.plot_components(fcmin)
```

```python
plt.plot(fcmin['ds'].tail(30), fcmin['yhat'].tail(30))
plt.plot(fcmax['ds'].tail(30), fcmax['yhat'].tail(30))
```
